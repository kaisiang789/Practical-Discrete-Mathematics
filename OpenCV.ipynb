1) load any creative common image from wikipedia.Enlarge the image double of its size

```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# URL of the Creative Commons image on Wikipedia
image_url = "https://upload.wikimedia.org/wikipedia/en/9/96/SatoruGojomanga.png"

# Download the image
response = requests.get(image_url)
img = Image.open(BytesIO(response.content))

# Double the size of the image
new_size = (img.width * 2, img.height * 2)
enlarged_img = img.resize(new_size)

# Save the enlarged image
enlarged_img.save("SatoruGojomanga1.png")

print("Image has been enlarged and saved as 'SatoruGojomanga.png'.")

fig, axes = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns
axes[0].imshow(img)
axes[0].set_title("Original Image")
axes[0].axis('off')  # Hide axes

axes[1].imshow(enlarged_img)
axes[1].set_title("Enlarged Image")
axes[1].axis('off')  # Hide axes

plt.show()
```

2 transpose the image 90 degree

```python
from PIL import Image
import matplotlib.pyplot as plt

# ... (your existing code) ...

# Load the enlarged image
enlarged_img = Image.open("SatoruGojomanga.png")

# Transpose the image 90 degrees clockwise
transposed_img = enlarged_img.transpose(Image.ROTATE_90)

# Save the transposed image
transposed_img.save("SatoruGojomanga_transposed.png")

# Display the transposed image (optional)
plt.imshow(img)
plt.axis('off')  # Hide axes
plt.show()

plt.imshow(transposed_img)
plt.axis('off')  # Hide axes
plt.show()
```
