1) load any creative common image from wikipedia.Enlarge the image double of its size

```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# URL of the Creative Commons image on Wikipedia
image_url = "https://upload.wikimedia.org/wikipedia/en/9/96/SatoruGojomanga.png"


# Download the image
response = requests.get(image_url)
img = Image.open(BytesIO(response.content))
img.save('SatoruGojomanga.png')

# Double the size of the image
new_size = (img.width * 2, img.height * 2)
enlarged_img = img.resize(new_size)

# Save the enlarged image
enlarged_img.save("SatoruGojomanga1.png")

print("Image has been enlarged and saved as 'SatoruGojomanga.png'.")

fig, axes = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns
axes[0].imshow(img)
axes[0].set_title("Original Image")
axes[0].axis('off')  # Hide axes

axes[1].imshow(enlarged_img)
axes[1].set_title("Enlarged Image")
axes[1].axis('off')  # Hide axes

plt.show()
```

Find 2 creative common images online. Combine these 2 images side by side vertically. Export it as Jpg

```python
import requests
from io import BytesIO
from PIL import Image
import matplotlib.pyplot as plt

# URLs of the Creative Commons images
image_url1 = "https://upload.wikimedia.org/wikipedia/en/9/96/SatoruGojomanga.png"  # Example URL
image_url2 = "https://upload.wikimedia.org/wikipedia/en/c/c8/Doraemon_volume_1_cover.jpg"  # Example URL

# Download and save the images
response = requests.get(image_url1)
img1 = Image.open(BytesIO(response.content)) # Assign to img1
img1.save('image1.png') #Save the first image

response = requests.get(image_url2)
# Check if the request was successful
if response.status_code == 200:
    img2 = Image.open(BytesIO(response.content)) # Assign to img2
    img2.save('image2.jpg') #Save the second image
else:
    print(f"Failed to download image from {image_url2}. Status code: {response.status_code}")


print("Images downloaded and saved as image1.png and image2.jpg.") # Corrected file extension for image1
# Open the images
image1 = Image.open("image1.png") # Corrected file extension to .png
image2 = Image.open("image2.jpg")

# Get the total width and height
total_width = max(image1.width, image2.width)
total_height = image1.height + image2.height

# Create a new image with the combined size
new_image = Image.new('RGB', (total_width, total_height))

# Paste the images onto the new image
new_image.paste(image1, (0, 0))  # Top image
new_image.paste(image2, (0, image1.height))  # Bottom image

# Save the combined image as a JPG
new_image.save("combined_image.jpg")

plt.imshow(image1)
plt.axis('off')  # Hide axes
plt.show()

plt.imshow(image2)
plt.axis('off')  # Hide axes
plt.show()

combined_image = Image.open("combined_image.jpg")
plt.imshow(combined_image)
plt.axis('off')  # Hide axes
plt.show()
```
