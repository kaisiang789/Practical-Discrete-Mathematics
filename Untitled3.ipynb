```python
# 安装必要的库
!pip install folium requests

import folium
import requests

# 第一步：下载马来西亚的GeoJSON数据
url = "https://raw.githubusercontent.com/johan/world.geo.json/master/countries/MYS.geo.json"
response = requests.get(url)
malaysia_geojson = response.json()

# 第二步：创建一个以马来西亚为中心的地图
malaysia_map = folium.Map(location=[4.2105, 101.9754], zoom_start=6)

# 第三步：使用GeoJSON数据高亮显示马来西亚
folium.GeoJson(
    malaysia_geojson,
    name="Malaysia",
    style_function=lambda feature: {
        "fillColor": "yellow",    # 高亮颜色
        "color": "black",       # 边界颜色
        "weight": 3,            # 边界厚度
        "fillOpacity": 0.5      # 透明度
    }
).add_to(malaysia_map)

# 第四步：添加图层控制（可选）
folium.LayerControl().add_to(malaysia_map)

# 显示地图
malaysia_map
```
